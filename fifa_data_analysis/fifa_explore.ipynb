```python
import sqlite3
import pandas as pd
from pandas.api.types import infer_dtype
pd.set_option("display.max_rows", 200)

con = sqlite3.connect("../fifa_api_server/fifa.db")
cur = con.cursor()
qry = cur.execute("select * from players limit 5;")
df = pd.DataFrame(qry.fetchall(), columns=[i[0] for i in qry.description])

df
```

```python
def get_metadata(df):
    return pd.DataFrame({
        "colname": [i for i in df.columns],
        "count": [len(df[i]) for i in df.columns],
        "nulls": [df[i].isnull().sum() for i in df.columns],
        "dtype": [infer_dtype(df[i]) for i in df.columns]
    })


meta = get_metadata(df)

meta
```
